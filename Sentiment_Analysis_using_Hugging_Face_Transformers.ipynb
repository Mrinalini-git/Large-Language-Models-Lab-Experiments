```python
from transformers import pipeline
sentiment_pipeline = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
user_text_for_sentiment = input("Enter a sentence for sentiment analysis: ")

# Perform sentiment analysis
sentiment_result = sentiment_pipeline(user_text_for_sentiment)

print(f"\nOriginal Sentence: {user_text_for_sentiment}")
print(f"Sentiment Result: {sentiment_result}")
print(f"Prediction: {sentiment_result[0]['label']} with score {sentiment_result[0]['score']:.4f}")
```
